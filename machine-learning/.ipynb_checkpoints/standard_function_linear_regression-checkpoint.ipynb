```python
import numpy as np 
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
# read data
data = np.genfromtxt("data.csv", delimiter=",")
x_data = data[:,0,np.newaxis]
y_data = data[:,1,np.newaxis]
plt.scatter(x_data, y_data)
plt.show()
```

```python
print(np.mat(x_data).shape)
print(np.mat(y_data).shape)
X_data = np.concatenate((np.ones((100,1)),x_data),axis=1)# 对数据进行处理 连接添加100行1 在第一列.
print(X_data.shape)
print(X_data[:3])
```

```python
# standard function for parameter of regression
def weights(xArr,yArr):
    xMat = np.mat(xArr)# 数组转化为矩阵
    yMat = np.mat(yArr)
    xTx = xMat.T*xMat
    if np.linalg.det(xTx) == 0.0:# 计算矩阵的值,如果为0 说明该矩阵没有inverse matrix
        print("the matrix cannot do inverse")
        return 
    ws = xTx.I*xMat.T*yMat  # xTx.I 是xTx inverse metrix
    return ws
```

```python
ws = weights(X_data, y_data)
print(ws)
```

```python
# draw result
x_test = np.array([[20],[75]])  # x 的取值范围
y_test = ws[0] + x_test * ws[1]
plt.plot(x_data, y_data, "b.")
plt.plot(x_test, y_test, "r")
plt.show()
```
