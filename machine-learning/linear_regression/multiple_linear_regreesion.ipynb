```python
import numpy as np 
from numpy import genfromtxt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # 画三维图像
```

```python
# read data 
data = genfromtxt(r"diliver.csv", delimiter=",")# 多元线性回归,画出的是一个平面,这里是三元
print(data)
```

```python
# 切分数据slice data 
x_data = data[:,:-1] # 从第0行到最后 :-1 是从第0列到最后一列但是不包括最后一列切分出来
y_data = data[:,-1]# 行从头到尾 列只切分出最后一列
print(x_data)
print(y_data)
```

```python
lr = 0.0001 # learning rate
# parameters
theta0 = 0
theta1 = 0
theta2 = 0
# maximum interation
iteration = 1000

def computer_error(theta0, theta1, theta2, x_data, y_data):
    totalError = 0
    for i in range(0,len(x_data)):
        totalError += (y_data[i] - (theta1 * x_data[i,0] + theta2 * x_data[i,1] + theta0)) **2
    return totalError / float(len(x_data))

def gradient_descent_runner(x_data, y_data, theta0, theta1, theta2, lr, iteration):
    m = float(len(x_data))
    for i in range (iteration):
        theta0_grad = 0
        theta1_grad = 0
        theta2_grad = 0
        # 
        for j in range(0, len(x_data)):
            theta0_grad += -(1/m) * (y_data[j] - (theta1 * x_data[j,0] + theta2 * x_data[j,1] + theta0))
            theta1_grad += -(1/m) * x_data[j,0] * (y_data[j] - (theta1 * x_data[j,0] + theta2 * x_data[j,1] + theta0))
            theta2_grad += -(1/m) * x_data[j,1] * (y_data[j] - (theta1 * x_data[j,0] + theta2 * x_data[j,1] + theta0))
        theta0 = theta0 - (lr * theta0_grad)
        theta1 = theta1 - (lr * theta1_grad)
        theta2 = theta2 - (lr * theta2_grad)
        return theta0, theta1, theta2
```

```python
print("Starting theta0 = {0}, theta1= {1}, theta2={2},error= {3}".format(theta0, theta1, theta2,computer_error(theta0, theta1,theta2, x_data, y_data)))
print("running...")
theta0, theta1,theta2,= gradient_descent_runner(x_data, y_data, theta0, theta1,theta2, lr, iteration)
print("after {0} iteration theta0={1}, theta1={2}, theta2={3}, error = {4}".format(iteration, theta0, theta1, theta2,computer_error(theta0,theta1,theta2, x_data, y_data)))
```

```python
# 图表的形式展示:
ax = plt.figure().add_subplot(111, projection = "3d")
ax.scatter(x_data[:,0], x_data[:,1], y_data, c = "r", marker = "o", s = 100)# 
x0 = x_data[:,0]
x1 = x_data[:,1]
# 生成网格矩阵
x0, x1 = np.meshgrid(x0,x1) 
z = theta0 + x0*theta1 + x1* theta2 
# 画3d图
ax.plot_surface(x0, x1, z)
ax.set_xlabel("Miles")
ax.set_ylabel("num of deliverise")
ax.set_zlabel("time")
plt.show()
```
