```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import make_gaussian_quantiles
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 生成2D正态分布, 生成的数据按分位数分为两类,500个样本, 2个样本特征
# 可以生成两类或者多类数据
x_data, y_data = make_gaussian_quantiles(n_samples = 500, n_features = 2, n_classes=2)
plt.scatter(x_data[:, 0], x_data[:,1],c=y_data)
plt.show()
```

```python
logistic = linear_model.LogisticRegression() # 这里默认使用的线性逻辑regression
logistic.fit(x_data, y_data)
```

```python
# 获取数据所在的范围, 画出网格矩阵
x_min, x_max = x_data[:,0].min()-1, x_data[:,0].max()+1
y_min, y_max = x_data[:,1].min()-1, x_data[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max,0.02))
z = logistic.predict(np.c_[xx.ravel(), yy.ravel()])# ravel and flatten are similar , 多维数据转1D ravel 改变原始数据
z = z.reshape(xx.shape)
# 等高线
cs = plt.contourf(xx, yy, z)
# 样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()
print("score", logistic.score(x_data, y_data))
print(z)
```

```python
# 定义多项式回归, degree的值可以调节多样式的特征
poly_reg = PolynomialFeatures(degree = 5)
x_poly = poly_reg.fit_transform(x_data)
# define regression model 
logistic = linear_model.LogisticRegression()
# training model 
logistic.fit(x_poly, y_data)
# data range , and net matrix 
x_min, x_max = x_data[:,0].min()-1, x_data[:,0].max()+1
y_min, y_max = x_data[:,1].min()-1, x_data[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max,0.02))
z = logistic.predict(poly_reg.fit_transform(np.c_[xx.ravel(), yy.ravel()]))
z = z.reshape(xx.shape)
# 等高线
cs = plt.contourf(xx, yy, z)
# 样本散点图
plt.scatter(x_data[:,0],x_data[:,1],c=y_data)
plt.show()
print("score", logistic.score(x_poly, y_data))
```
