```python
"""
bagging 思想就是在数据集中 放回抽取样本作为训练样本 构建多个模型, 多个模型进行打分测试模型 得出一个结果
"""
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import neighbors
from sklearn import datasets
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
# read data 
iris = datasets.load_iris()
print(iris.data[:3]) # iris data are four culm feature 
x_data = iris.data[:,:2] #  我们只使用两个特征 因为我们需要比较低的准确率 使用bagging清楚看到提升
y_data = iris.target
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data)
```

```python
# building model :knn 
knn = neighbors.KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
def plot(model):
    # 获取值的范围
    x_min, x_max = x_data[:,0].min()-1, x_data[:,0].max()+1
    y_min, y_max = x_data[:,1].min()-1, x_data[:,1].max()+1
    # 生成网络矩阵 画图背景
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
    z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    cl = plt.contourf(xx, yy, z)
```

```python
# drawing 
plot(knn)
plt.scatter(x_data[:,0], x_data[:,1], c=y_data)
plt.show()
knn.score(x_test, y_test)
```

```python
# building decision tree model 
dtree = tree.DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

```python
plot(dtree)
plt.scatter(x_data[:,0], x_data[:,1], c=y_data)
plt.show()
dtree.score(x_test, y_test)
```

```python
# Bagging_KNN 装袋
bagging_knn = BaggingClassifier(knn, n_estimators = 100)
# input data building model 
bagging_knn.fit(x_train, y_train)
plot(bagging_knn)
plt.scatter(x_data[:,0], x_data[:,1], c= y_data)
plt.show()
bagging_knn.score(x_test, y_test)
```

```python
# bagging_decisionTree 
bagging_tree = BaggingClassifier(dtree, n_estimators = 100)
bagging_tree.fit(x_train, y_train)
plot(bagging_tree)
plt.scatter(x_data[:,0], x_data[:,1], c = y_data)
plt.show()
bagging_tree.score(x_test, y_test)
```
